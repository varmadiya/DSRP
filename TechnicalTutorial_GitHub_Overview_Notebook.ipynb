## **Part 1: What is a Repository?**
---

A **repository** (repo) is a project's home on GitHub. It holds your files along with a history of every saved version, so you can always look back at how your files have been changed and the refer back to past versions.

A **commit** is one saved version within your repo's version history. Each commit comes with a short message that describes the changes made.


---
## **Part 2: Create a Repository on GitHub**
---

1. Go to [github.com/new](https://github.com/new) (log in first if needed).
2. Give it a name, like `dsrp`.
3. You can leave it **Public** or set it to **Private**, does not matter for this notebook.
4. Leave everything else as-is (don't add a README) and click **Create repository**.

---
## **Part 3: Connect Colab to GitHub and Save This Notebook**
---

1. In the Colab menu bar, click **File → Save a copy in GitHub**.
2. The first time you do this, Colab will ask permission to access your GitHub account — click **Authorize**.
3. A dialog box will appear:
   - **Repository**: choose the one you just created.
   - **Branch**: leave this as `main`.
   - **File path**: this can stay as the notebook's current name.
   - **Commit message**: write a short note, like `Initial save of workshop notebook`.
4. Click **OK**. Colab will open a new tab showing your notebook now living on GitHub.

Go check your GitHub repository page — you should see your notebook file listed there.


---
## **Part 4: Make a Change and Save a New Version**
---

Every time you save a copy to GitHub with a new commit message, you're adding another entry to your project's history — you never lose the earlier versions.

Edit the code cell below (change the message, add a line, anything), then run it.


```
# Edit this however you like, then run this cell
message = "Hello from my first GitHub commit!"
print(message)
```

Now save your progress again:

1. **File → Save a copy in GitHub**.
2. Same repository, same branch — but this time write a **new** commit message describing what you changed, like `Updated greeting message`.
3. Click **OK**.

This time, Colab is saving a *new version* to the *same* file, not creating a duplicate.


---
## **Part 5: See Your History on GitHub**
---

1. Go to your repository on GitHub.
2. Click on your notebook's filename to open it.
3. Click **History** (near the top of the file view).

You'll see every commit you've made, in order, each with the message you wrote. This is the version history of your project — every research notebook you save from here forward will build up its own timeline just like this.


---
## **Part 6: Reopening a Notebook From GitHub**
---

If you come back later (a new day, a different computer) and want to pick up where you left off:

1. In Colab, go to **File → Open notebook**.
2. Click the **GitHub** tab.
3. Search for your repository, and select the notebook file.

Colab will load the latest saved version straight from your repository.


---
## Recap: Glossary
---

| Term | Meaning |
|---|---|
| **Repository** | A project's home on GitHub — holds your files and their full saved history. |
| **Commit** | One saved checkpoint of your work, with a short message describing what changed. |
| **Push** | What happens behind the scenes when you click "Save a copy in GitHub" — your changes are uploaded to the repository. |
